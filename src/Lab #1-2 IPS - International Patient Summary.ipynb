```java
import ca.uhn.fhir.context.FhirContext;
import ca.uhn.fhir.rest.client.api.*;
import ca.uhn.fhir.rest.api.MethodOutcome;
import ca.uhn.fhir.validation.*;
import org.hl7.fhir.r4.model.*;
import java.nio.file.*;
```

```java
FhirContext ctx = FhirContext.forR4();
// String serverBase = "https://hapi.fhir.org/baseR4";
String serverBase = "http://localhost:8080/fhir";
ctx.getRestfulClientFactory().setSocketTimeout(60 * 10 * 100);
IGenericClient client = ctx.newRestfulGenericClient(serverBase);
```

```java
//read all Patient from Server using Get /Patient
Bundle response = client.search().forResource(Patient.class).returnBundle(Bundle.class).execute();

//print all Patient id
for (Bundle.BundleEntryComponent entry : response.getEntry()) {
    System.out.println(entry.getResource().getId());
}
```

```java
// get the summary for a single patient
Bundle bundle = client
    .operation()
    .onInstance("Patient/997")
    .named("$summary")
    .withNoParameters(Parameters.class)
    .returnResourceType(Bundle.class)
    .execute();
    
//print the raw json    
System.out.println(ctx.newJsonParser().setPrettyPrint(true).encodeResourceToString(bundle));
```
